```python
# Importando bibliotecas
import pandas as pd
import numpy as np
```

```python
# Importando dataset e lendo as primeiras colunas
houses_sp = pd.read_csv('/houses_sp.csv')
houses_sp.head()
```

```python
# Verificando distribuição de imóveis no dataset
houses_sp['city'].value_counts()
```

```python
# Puxando todas as informações do dataset
houses_sp.info()
```

Observando que a coluna cidade é irrelevante , pois existe apenas a cidade de SP a mesma pode ser descartada.


```python
# Excluindo coluna cidade (dado constante)
houses_sp = houses_sp.drop(columns=['city'])
```

```python
# Lendo df alterado
houses_sp.head()
```

Verificando dados errôneos

```python
# Verificando dados faltantes na coluna de quartos
houses_sp['rooms']
```

```python
# Calculando a mediana de quartos para preencher os campos nulos
houses_sp['rooms'].median()
```

```python
# Preenchendo os campos com a mediana
houses_sp['rooms'].fillna(houses_sp['rooms'].median(), inplace=True)
```

```python
# Verificando dados apos preenchimento na coluna de quartos
houses_sp['rooms']
```

Verificando a coluna andar observa-se alguns dados faltantes que podem ser substituidos por um dado coerente apos verificar sua frequência

```python
# Verificando a frequência de dados faltantes na coluna andar
houses_sp['floor'].value_counts()
```

```python
# Filtrando os 1847 dados faltantes
houses_sp.loc[houses_sp['floor'] == '-', 'floor'] = '0'
# Verificando a alteração
houses_sp.head()
```
